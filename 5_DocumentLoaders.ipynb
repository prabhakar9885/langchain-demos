```python
import os
from constants import openai_key

os.environ['OPENAI_API_KEY'] = openai_key
```

```python
from PyPDF2 import PdfReader

pdf_reader = PdfReader("/Users/prabhakar/Downloads/budget_speech.pdf");
```

```python
raw_text = ""

for page in pdf_reader.pages:
    if page.extract_text():
        raw_text += page.extract_text()
```

```python
print( raw_text )
```

```python
from langchain_text_splitters import CharacterTextSplitter

characterTextSplitter = CharacterTextSplitter(
    separator="\n",
    chunk_size=1000, 
    chunk_overlap=200, 
    length_function=len
)

texts = characterTextSplitter.split_text(raw_text)

print( len(texts) )
```

```python
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings()
faiss = FAISS.from_texts( texts, embeddings )
```

```python
from langchain_openai import OpenAI
from langchain.chains.question_answering import load_qa_chain

qa_chain = load_qa_chain(
    OpenAI(), 
    chain_type="stuff"
)
```

```python
query="Are tax payers provided with any relief?"

qa_chain.invoke( {
        "input_documents" : faiss.similarity_search(query=query),
        "question" : query
    }
)
```

```python
query="What benefits are given to agriculturalists ?"

qa_chain.invoke( {
        "input_documents" : faiss.similarity_search(query=query),
        "question" : query
    }
)
```
